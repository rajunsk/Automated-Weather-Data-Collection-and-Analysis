The dataset contains daily weather observations consisting of 4081+ records and multiple weather-related variables. Each record represents the weather conditions for a specific day and includes information such as temperature, rainfall, wind speed, and derived temporal features.

**Working with Date and Time**

```python
import datetime
```

```python
data['date'] = pd.to_datetime(data['date'])
data["month"] = data['date'].dt.month
```

**Creating Seasonal Categories**

```python
def get_seasons_month(month):
    if month in [12,1,2]:
        return "winter"
    elif month in [3,4,5]:
        return "summer"
    elif month in [6,7,8]:
        return "monsoon"
    else:
        return "autumn"
data["season"] = data["month"].apply(get_seasons_month)
```
